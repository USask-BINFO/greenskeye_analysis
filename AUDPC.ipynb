```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import cv2
import skimage.io as io
```

```python
"""
Basic function for calculating Area Under the Disease Precision Curve using one measure per day

Inputs:
current_df: the dataframe where you store the filenames and infection data from rust_analysis_workflow.ipynb
start_date: the first day you want to measure AUDPC from, as an int
end_date: the last day you want to measure AUDPC from, as an int
label: the string label in your dataframe of disease measurement ('%_infected' or 'Pustule Count')

returns: infection at each time point used to calculate AUDPC, the calculated AUDPC using the trapezoidal rule

other: reading in filenames from the dataframe is specific to the greenskeye data management format, may need to change for your work
"""

def AUDPC(current_df, start_date, end_date, measurements_per_day, label):
    filepaths = current_df.index.tolist()
    dates = []
    # note that this is specific to our form of file organization
    for entry in filepaths:
        entry = entry.split('/')
        filename = entry[-1]
        filename = filename.split('_')
        dates.append(filename[-3])
    
    start_index = dates.index(str(start_date))
    end_index = dates.index(str(end_date))
    end_time = max(i for i, d in enumerate(dates) if d == str(end_date))
    
    audpc_samples = current_df[start_index:end_index:measurements_per_day] 

    audpc_samples = pd.concat([audpc_samples, current_df.iloc[[end_time]]])
    # calculate AUDPC
    # AUDPC
    disease_progression = audpc_samples[label].to_list()
    audpc = np.trapezoid(disease_progression)

    return audpc, disease_progression
```
